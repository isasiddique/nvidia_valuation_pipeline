```python
import os
import numpy as np
import pandas as pd

# 1. CREATE DATA ARCHITECTURE FOLDER
os.makedirs('data', exist_ok=True)

# Set seed for exact scientific reproducibility
np.random.seed(42)
n_simulations = 2000

# Base Financial Truth (2026 Baseline Metrics)
current_revenue_bn = 253.5   # Trailing 12-Month Revenue in Billions
base_net_margin = 0.55       # ~55% net profit margins
base_pe_multiple = 31.0      # Baseline Price-to-Earnings Ratio

# 2. RUN MONTE CARLO PROJECTIONS FOR THE YEAR 2028
simulated_market_caps = []
simulated_revenues = []

for _ in range(n_simulations):
    # Constraint A: Hyperscaler Cloud CapEx growth rates vary from 15% to 54% annually
    annual_growth_rate = np.random.uniform(0.15, 0.54)
    projected_rev_2028 = current_revenue_bn * ((1 + annual_growth_rate) ** 2)

    # Constraint B: Competitor chip pressure degrades hardware margins slightly
    simulated_margin = base_net_margin - np.random.uniform(0.0, 0.08)
    projected_earnings = projected_rev_2028 * simulated_margin

    # Constraint C: Market sentiment volatility contracts or expands the P/E multiple
    simulated_pe = base_pe_multiple * np.random.uniform(0.70, 1.15)

    # Calculate Projected Market Capitalization
    market_cap_projection = (projected_earnings * simulated_pe)
    simulated_market_caps.append(market_cap_projection)
    simulated_revenues.append(projected_rev_2028)

# 3. CREATE SIMULATION DATAFRAME AND SAVE TO REPOSITORY
df_finance = pd.DataFrame({
    'SimulationID': [f"SIM-{i:04d}" for i in range(1, n_simulations + 1)],
    'ProjectedRevenue_BN': np.round(simulated_revenues, 2),
    'ProjectedMarketCap_BN': np.round(simulated_market_caps, 2)
})

df_finance.to_csv('data/raw_nvidia_projections.csv', index=False)

# 4. PRINT DISTRIBUTION STATISTICS
median_cap = np.median(simulated_market_caps)
bear_case = np.percentile(simulated_market_caps, 10)  # 10th percentile
bull_case = np.percentile(simulated_market_caps, 90)  # 90th percentile

print("--- QUANTITATIVE VALUATION ANALYSIS COMPLETE ---")
print(f"Projected Median 2028 Market Cap: ${median_cap:.2f} Billion")
print(f"Bear Case Valuation (10th Percentile): ${bear_case:.2f} Billion")
print(f"Bull Case Valuation (90th Percentile): ${bull_case:.2f} Billion")
```

```python
# 1. LOAD PROJECTIONS DATA
df_results = pd.read_csv('data/raw_nvidia_projections.csv')

# 2. EVALUATE MARKET SEGMENT RISKS
print("--- RISK METRICS & PROBABILITY OF VALUE VARIANCE ---")
prob_below_current = (df_results['ProjectedMarketCap_BN'] < 4910.0).mean() * 100
print(f"Probability of Market Cap contracting below current $4.91T: {prob_below_current:.1f}%")

avg_revenue_bull = df_results[df_results['ProjectedMarketCap_BN'] >= bull_case]['ProjectedRevenue_BN'].mean()
print(f"Average 2028 Corporate Revenue required to hit Bull Case: ${avg_revenue_bull:.1f} Billion")
```

## 📘 Code Explainer: Monte Carlo Financial Engine
*This document translates the mathematical simulation code into universal language for non-technical stakeholders.*

### 1. Modeling Financial Uncertainty
* **`np.random.uniform(0.15, 0.54)`**: Instead of using a single biased growth percentage guess, this tells the simulator that annual cloud hardware data center infrastructure spending could realistically range between a 15% floor and a 54% ceiling.
* **`current_revenue_bn * ((1 + annual_growth_rate) ** 2)`**: This uses compound growth math to calculate where NVIDIA's base revenue will land over a 2-year timeline under 2,000 distinct parallel realities.
* **`base_net_margin - np.random.uniform(0.0, 0.08)`**: This simulates real-world industry friction. We subtract up to 8% from corporate profit margins to represent competitors creating their own internal custom tensor processing silicon chips, forcing down prices.

### 2. Slicing Out Volatility
* **`np.percentile(..., 10)` & `np.percentile(..., 90)`**: Financial analysts use percentiles to manage risk. The 10th percentile calculates the "Bear Case" (what happens if all negative macro constraints hit at once), while the 90th percentile maps the "Bull Case" (maximum market upside).
